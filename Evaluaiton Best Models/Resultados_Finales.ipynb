## DAX (mejor entrenamiento)

```python
# Se cargan los datos 
yahoo_dfDAX = pd.read_csv("Yahoo_DAX_Final.csv")
fecha_inicialDAX = '2010-01-18'
fecha_finalDAX = '2024-01-18'


# Mejores hiperparámetros en entrenamientos previos
len_window_DAX = 10
n_epochs_DAX = 500
input_size=1
num_layers_DAX = 1
hidden_size_DAX= 4
learning_rate_DAX = 0.000045
escalamientoDAX = MinMaxScaler
loss_fn = nn.MSELoss()

# Se crea el dataset y la división según la ventana temporal
dataSetDAX = TimeSeries_Dataset(yahoo_dfDAX, fecha_inicialDAX, fecha_finalDAX, len_window_DAX)

# Se carga el modelo
modelDAX = LSTM(input_size, hidden_size_DAX, num_layers_DAX)
optimizerDAX = torch.optim.Adam(modelDAX.parameters(), learning_rate_DAX)
entrenamientoDAX = ModelTraining(modelDAX, loss_fn, optimizerDAX)
entrenamientoDAX.set_loaders(dataSetDAX.train_loader, dataSetDAX.test_loader)

# Se entrena el modelo 
entrenamientoDAX.train(n_epochs_DAX)
```

```python
# Se visualizan el entrenamiento 
#Se Grafican los resultados
fig1_1 = entrenamientoDAX.plot_MSE(Title="Training: 500 epochs, lr = 0.00045, layers=1 (DAX)")
fig2_1 = entrenamientoDAX.plot_forecasting(dataSetDAX.X_train , dataSetDAX.y_train, dataSetDAX.train_dates ,Title="Train data set (DAX)", train_or_test = 'train')
fig3_1 = entrenamientoDAX.plot_forecasting(dataSetDAX.X_test,dataSetDAX.y_test, dataSetDAX.test_dates, Title='Test Data set (DAX)', train_or_test = 'test')

# Se calculan las métricas
metricasDAX = entrenamientoDAX.compute_metrics(X_train=dataSetDAX.X_train.numpy(), y_train=dataSetDAX.y_train.numpy(),
                                 X_test=dataSetDAX.X_test.numpy(), y_test=dataSetDAX.y_test.numpy())

# Se imprimen las métricas
metricasDAX
```
