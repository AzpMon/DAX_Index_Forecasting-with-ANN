# Creaci贸n del DataSet (pytorch)

```python
#Importaci贸n de las librerias
import pandas as pd
from TimeSeries_Dataset import *

yahoo_df = pd.read_csv("Yahoo_DAX_Final.csv")
fecha_inicial = '2010-01-18'
fecha_final = '2024-01-18'
len_window = 10

#Creaci贸n del Dataset 
dataSet = TimeSeries_Dataset(yahoo_df, fecha_inicial, fecha_final, len_window)
```

# Entrenamiento  del modelo LSTM+RNN-FC a distintas capas

```python
#Se importan las librerias
from LSTM_Class import *
from ModelTraining import *
import torch.nn as nn


#Funci贸n de costo y optimizador del Modelo
loss_fn = nn.MSELoss()
learning_rate = 0.00045
```

### Prueba 1

```python
model1 = LSTM(input_size=1, hidden_size=4, num_layers=1)
optimizer1 = torch.optim.Adam(model1.parameters(), learning_rate)
entrenamiento1 = ModelTraining(model1, loss_fn, optimizer1)
entrenamiento1.set_loaders(dataSet.train_loader, dataSet.test_loader)

#Se entrena al modelo
entrenamiento1.train(n_epochs=500)

#Se Grafican los resultados
fig1_1 = entrenamiento1.plot_MSE(Title="Training: 500 epochs, lr = 0.00045, layers=1")
fig2_1 = entrenamiento1.plot_forecasting(dataSet.X_train , dataSet.y_train, dataSet.train_dates ,Title="Train data set", train_or_test = 'train')
fig3_1 = entrenamiento1.plot_forecasting(dataSet.X_test,dataSet.y_test, dataSet.test_dates, Title='Test Data set', train_or_test = 'test')
```
